```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns
from sklearn.linear_model import LinearRegression

# 设置中文显示和字体大小
plt.rcParams['font.sans-serif'] = ['Times new roman']
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.size'] = 14
```

```python
data = pd.read_csv(r"C:\Users\Eugene\Desktop\results.csv")
# 将列名中的空格去除
data.columns = data.columns.str.strip()

# 获取 epoch 的最后一个数
epoch_str = str(data['epoch'].iloc[-1])
epoch_last_num = epoch_str.split('.')[-1]

# 定义子图标题和数据列名
titles = [['Train Box Loss', 'Train Object Loss', 'Train Class Loss', 'Validation Box Loss', 'Validation Object Loss'],
          ['Validation Class Loss', 'Precision', 'Recall', 'mAP@0.5', 'mAP@0.5:0.95']]
columns = [['train/box_loss', 'train/obj_loss', 'train/cls_loss', 'val/box_loss', 'val/obj_loss'],
           ['val/cls_loss', 'metrics/precision', 'metrics/recall', 'metrics/mAP_0.5', 'metrics/mAP_0.5:0.95']]

# 绘制图表
fig, axs = plt.subplots(2, 5, figsize=(20, 10))
fig.suptitle(f'Epoch: {epoch_last_num}', fontsize=24)

# 循环调用子图绘制函数
for i in range(2):
    for j in range(5):
        ax = axs[i, j]
        col = columns[i][j]
        title = titles[i][j]
        ax.plot(data['epoch'], data[col], linewidth=2)
        ax.set_title(title)
        ax.set_xlabel('Epoch', fontsize=16)
        if j == 0:
            ax.set_ylabel('Loss / Metric', fontsize=16)
        ax.grid(True, linestyle='--', linewidth=1, alpha=0.5)
        for spine in ['top', 'right']:
            ax.spines[spine].set_visible(False)

# 调整子图的间距
plt.subplots_adjust(wspace=0.15, hspace=0.3)

# plt.savefig(f'Epoch：{epoch_last_num}.svg', format='svg')
plt.show()
```
